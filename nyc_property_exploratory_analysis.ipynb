## Data Cleaning
The column `"Unnamed: 0"` just iterates by 1 and provides no other information, it can be removed.

```python
property_sales.drop("Unnamed: 0", inplace=True, axis=1)
property_sales.info()
```

The column `"EASE-MENT"` does not look like it has any information in it, let's see a list of the unique values in it.

```python
property_sales["EASE-MENT"].unique()
```

It only has blank values, so it can also be removed.

```python
property_sales.drop("EASE-MENT", inplace=True, axis=1)
```

The dataset is using "-" and " " to represent missing values. These should be replaced with nans.

```python
property_sales.replace({"-":np.nan, " ":np.nan}, inplace=True)
```

```python
property_sales.isnull().sum()
```

### Fixing column types

```python
property_sales.info()
```

The columns `"LAND SQUARE FEET"`, `"GROSS SQUARE FEET"`, and `"SALE PRICE"` should be entirely numeric values, but their data types are not integers. Any non-integer values in these columns should be converted to nan and the column converted to type `float64`. `errors` will be set to coerce to make all non-numeric values nans.

```python
for col in ["LAND SQUARE FEET", "GROSS SQUARE FEET", "SALE PRICE"]:
  property_sales[col] = pd.to_numeric(property_sales[col], errors="coerce")

property_sales.info()
```

The `"SALE DATE"` column can be converted to datetimes.

```python
property_sales["SALE DATE"] = pd.to_datetime(property_sales["SALE DATE"])
```
